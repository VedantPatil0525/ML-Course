```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices.csv")
```

```python
df.head()
```

```python
%matplotlib inline
plt.xlabel("Area(Sq Ft)")
plt.ylabel("Prices(USD$)")
plt.scatter(df.area, df.price, color='black', marker='*')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
%matplotlib inline
plt.xlabel("Area(Sq Ft)")
plt.ylabel("Prices(USD$)")
plt.scatter(df.area, df.price, color='black', marker='*')
plt.plot(df.area, reg.predict(df[['area']]), color='maroon')
```

```python
reg.predict([[5000]])
```

```python
# value of m, i.e., coefficient
reg.coef_
```

```python
# value of b, i.e., intercept
reg.intercept_
```

```python
# now we know, y=m*x+b, therefore
y = 135.78767123*5000+180616.43835616432
y
```
